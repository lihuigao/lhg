```python
import matplotlib
print(matplotlib.__version__) #查看Matplotlib版本
import pandas as pd
print(pd.__version__) #查看pandas版本
import numpy as np
print(np.__version__) #查看numpy版本
import matplotlib.pyplot as plt 
plt.rcParams['font.sans-serif'] = ['SimHei']  
```

```python
x = ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
y1 = np.array([58, 120, 86, 20, 139, 57, 139])
y2 = np.array([30,88,100,60,40,70,30])
```

```python
plt.figure(figsize=(10,7))#设置画布的尺寸

plt.bar(x = x,height = y1, label='系列1',alpha = 0.7,edgecolor = 'black')
plt.bar(x = x,height = -y2, label='系列2',alpha = 0.7,edgecolor = 'black')

plt.title('基础柱状图——正负柱状图',fontsize=20) #设置标题
plt.xlabel('我是X轴',fontsize=14) #设置x轴
plt.ylabel('我是Y轴',fontsize=14) #设置y轴

plt.legend(loc=2,fontsize=13)  # 设置图例位置

plt.show()
```

```python
menMeans = (20, 35, 30, 35, 27)  # 男生各组平均分
womenMeans = (25, 32, 34, 20, 25)# 女生各组平均分
menStd = (2, 3, 4, 1, 2)  # 男生组各标准差
womenStd = (3, 5, 2, 3, 3) # 女生组各标准差
label = ('第一组', '第二组', '第三种', '第四组', '第五组')
bar_width = 0.4
bar_x = np.arange(len(label)) 


fig = plt.figure(figsize=(8, 4))


ax = fig.add_subplot(111)
ax.set_title('图1 垂直柱状图')
bar1 = ax.bar(x=bar_x - bar_width/2,   # 设置不同的x起始位置
              height=menMeans, width=bar_width)
bar2 = ax.bar(x=bar_x + bar_width/2,   # 设置不同的x起始位置
              height=womenMeans, width=bar_width,
        )

ax.set_xlabel('组别')
ax.set_ylabel('分数')
ax.set_title('各组不同性别分数')
ax.set_xticks(range(5))
ax.set_xticklabels(label)
ax.set_yticklabels(np.arange(0, 81, 10))
ax.legend((bar1, bar2), ('男生', '女生'))
```
